```python
# Importing relevant modules 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Reading data from 'insurance.csv' file and storing contents into a dataframe.
df = pd.read_csv('insurance.csv', encoding='unicode_escape')

# Displaying first 5 rows of dataframe.
df.head()
```

### The Effect of Age on Insurance Costs

```python
# Age is the independent variable so its values are stored in variable 'x'.
# Insurance Charges is the independent variable so its values are stored in variable 'y'.
x = df.iloc[:,0].values
y = df.iloc[:,-1].values

# Scatter plot showing relationship between age and charges is displayed.
sns.scatterplot(data=df, x=x, y=y).set(xlabel='Age', ylabel='Charges')
```

```python
# Reshaping the x-variable to allow compatibility.
X = df.iloc[:,0].values.reshape(-1,1)

# Fitting the data and making a prediction.
insurance_model = LinearRegression()
insurance_model.fit(X,y)
y_prediction = insurance_model.predict(X)
```

```python
# Plotting the original scatterplot along with a line of best fit.
sns.scatterplot(data=df, x=x, y=y).set(xlabel='Age',ylabel='Charges')
plt.plot(X, y_prediction)
```

```python
# Making a prediction on new data.
age_1 = [[35]]

print(f"A person of age {age_1} will have an insurance cost of {insurance_model.predict(age_1)}")
```

#### Observation:
As a persons age increases, their insurance charges will generally increase too.

### The Effects of BMI on Insurance Costs

```python
# BMI is the independent variable so its values are stored in variable 'x'.
# The insurance charges variable 'y' from before is used here.
x = df.iloc[:,2].values

# Scatterplot showing the relationship between BMI and insurance costs is displayed.
sns.scatterplot(data=df, x=x, y=y).set(xlabel='BMI',ylabel='Charges')
```

```python
# Making a prediction on new data.
bmi_1 = [[27]]

print(f"A person with a BMI of {bmi_1} will have an insurance cost of {insurance_model.predict(bmi_1)}")
```
